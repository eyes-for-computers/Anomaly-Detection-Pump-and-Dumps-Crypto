```python
def download(url, filename):
    with open(filename, 'wb') as f:
        response = requests.get(url, stream=True)
        total = response.headers.get('content-length')

        if total is None:
            f.write(response.content)
        else:
            downloaded = 0
            total = int(total)
            for data in response.iter_content(chunk_size=max(int(total/1000), 1024*1024)):
                downloaded += len(data)
                f.write(data)
                done = int(50*downloaded/total)
                sys.stdout.write('\r[{}{}]'.format('█' * done, '.' * (50-done)))
                sys.stdout.flush()
    sys.stdout.write('\n')
```

'''
Trades
All symbols are supported, the file format is:
<base_url>/data/spot/monthly/trades/<symbol_in_uppercase>/<symbol_in_uppercase>-trades-<year>-<month>.zip

e.g. the url BNBUSDT trades in 2019-01 is:
https://data.binance.vision/data/spot/monthly/trades/BNBUSDT/BNBUSDT-trades-2019-01.zip

All intervals are supported: 1m, 3m, 5m, 15m, 30m, 1h, 2h, 4h, 6h, 8h, 12h, 1d, 3d, 1w, 1mo

1mo is used instead of 1M to supprt non-case sensitive file systems.

Klines
All symbols are supported, the file format is:
<base_url>/data/spot/monthly/klines/<symbol_in_uppercase>/<interval>/<symbol_in_uppercase>-<interval>-<year>-<month>.zip

e.g. the url for BNBUSDT 1m klines for 2019-01 is:
https://data.binance.vision/data/spot/monthly/klines/BNBUSDT/1m/BNBUSDT-1m-2019-01.zip


'''


# want trades and klines csvs, those endpoints are:


```python
#Downlaod the Klines

def getKlines(symbol,period, day, hour):  #currenly in format YYYY-DD  
    
    
    '''
    
    getKlines() takes input of getSymbolData for a klines file and returns the proper url of the file
    
    '''
    
    symbol = str(symbol)
    base = 'BTC'
    pair = symbol + base

    baseKlinesURL = ['https://data.binance.vision/data/spot/monthly/klines/', pair, '/1m/', pair, '-1m-', str(period), '.zip']

    baseKlinesURL = ''.join(baseKlinesURL)
    
    
    filename = ['pump_klines_' , pair, '_', period, '-', day, '-', hour, '.zip']#.csv']
    filename = ''.join(filename)
    download(baseKlinesURL, filename)
    
    
    return print('done!', baseKlinesURL)
```

```python
# add the btc pair to each symbol

def getTrades(symbol,period, day, hour):
    
    '''
    
    getTrades() takes input of getSymbolData for a trades file and returns the proper url of the file
    
    '''
    
    
    symbol = str(symbol)
    base = 'BTC'
    pair = symbol + base

    baseTradesURL = ['https://data.binance.vision/data/spot/monthly/trades/', pair, '/', pair, '-trades-', str(period), '.zip']

    baseTradesURL = ''.join(baseTradesURL)
    
    
    
    filename = ['pump_trades_', pair, '_', period, '-', day, '-', hour, '.zip']
    filename = ''.join(filename)
    download(baseTradesURL, filename)
            
    return print('done!', baseTradesURL)

getTrades('FIL', '2021-05', '1', '1')
```

```python
from time import sleep

#here is where I add in the proper date?, no that even goes in the next fnction

def getSymbolData(symbol, period, day, hour):
    
    '''
    
    Input for the symbol you are looking for the time period to extract data from the fullPumpList 
    
    '''
    print('Downloading and writing out trades and Klines')
    getTrades(symbol,period, day, hour)
    getKlines(symbol, period, day, hour)
    sleep(3)
```
